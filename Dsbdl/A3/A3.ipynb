```python
# Import all the required Python Libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

# Part 1

```python
df=pd.read_csv('./nba.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['College']=df['College'].fillna(df['College'].mode())
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
df.dropna(inplace=True) #drop rows with null values     
df.isnull().sum()
```

```python
H_grp=df.groupby(df['Height'])
H_grp.head()
```

```python
H_grp.describe()
```

```python
bins=[19,25,31,36,40]
lables=['19-25','26-31','32-36','37-40']
df['A_grp']=pd.cut(df['Age'],bins=bins,labels=lables)

df['A_grp'].value_counts()
```
