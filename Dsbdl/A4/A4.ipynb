```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean())
df.dropna(inplace=True)
df.isnull().sum()
```

```python
sns.histplot(data=df, x='Age', kde=True) #kde is kernel density estimation and it is used to estimate the probability density function of a random variable
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(data=df, x="Survived", y="Age")
```

```python
sns.countplot(data=df, x="Embarked", hue="Survived") #hue is used to separate the data by a categorical variable
```

```python
sns.countplot(data=df, x="Sex", hue="Survived")
```
